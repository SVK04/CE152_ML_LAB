```python
import numpy as np
import torch

```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43],
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37],
                   [69, 96, 70], 
                   
                  ], dtype='float32')

# Targets (apples, oranges)
targets = np.array([[56],
                    [81],
                    [119], 
                    [22], 
                    [103],
                   
                    ],dtype='float32')



print(inputs)
print(targets)
```

```python
#Convert inputs and targets to tensors

inputs=torch.from_numpy(inputs)
print(inputs)
targets=torch.from_numpy(targets)
print(targets)
```

```python
#Weight and biases

w= torch.randn(1,3,requires_grad=True)
print(w)
b=torch.randn(1,requires_grad=True)
print(b)

```

```python
#Define The Model
def model(x):
  return torch.mm(x,w.t(),out=None)+b
```

```python
#Generate Predictions
#Compare with Targets

preds=model(inputs)
print("Prediction: ",preds)
print("Targets: ",targets)

```

```python
#MSE loss
#Compute Loss

def mse(P,T):
  diff=0
  i=0
  sum=0
  n=len(T)
  for k in range(0,n):
    i+=1
    diff=T[k]-P[k]
    square=diff*diff
    sum+=square
  mean=sum/i
  return mean

output=mse(preds,targets)
print("Loss: ",output)
```

```python
#Compute Gradients
print("Gradient :",output.backward())
```

```python
#Gradient for weights
print(w)
print(w.grad)
```

```python
#Gradient for bias
print(b)
print(b.grad)
```

```python
# Compute gradients
with torch.no_grad():
  w-=w.grad*1e-5
  b-=b.grad*1e-5
  w.grad.zero_()
  b.grad.zero_()
print(w)
print(b)
```

```python
# Generate predictions
# Calculate the loss
new_preds=model(inputs)
new_loss=mse(new_preds,targets)
print("New Loss: ",new_loss)

```

```python
#Train for 100 Epochs
for i in range(100):
  preds=model(inputs)
  loss=mse(preds,targets)
  loss.backward()
  with torch.no_grad():
     w-=w.grad*1e-5
     b-=b.grad*1e-5
     w.grad.zero_()
     b.grad.zero_()

# print(preds)
# print(loss)
```
