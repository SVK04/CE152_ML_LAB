```python
#4.2 Steps for Handling Categorical Data

from sklearn.preprocessing import LabelEncoder,OneHotEncoder

#Step2: Load Data

datasets=pd.read_csv('/content/drive/MyDrive/SEM-6/Ml/Lab2/Data_for_Categorical_Values.csv')
print(datasets)
datasets.describe()
```

```python
#Step 3

X=datasets.iloc[:,:-1].values
Y=datasets.iloc[:,-1].values

print(X)
print(Y)
```

```python
#Step 4a: Apply LabelEncoder on the data to convert 
#         country names into numeric values

le=LabelEncoder()
X[:,0]=le.fit_transform(X[:,0])
print(X)
```

```python
from pandas.io.sql import DatabaseError
#Step 4b: Use dummy variables from pandas library
#         to create one column for each country

dummy=pd.get_dummies(datasets['Country'])
print(dummy)
datasets=datasets.drop(['Country','Purchased'],axis=1)
datasets=pd.concat([dummy,datasets],axis=1)
print(datasets)
```

```python
#Use OneHotEncoder from scikit learn
onehotencoder=OneHotEncoder()
# reshape the 1-D cpuntry array to 2-D as fit_transform expects 2-D and
# finally fit the object

x=onehotencoder.fit_transform(datasets.Country.values.reshape(-1,1)).toarray()
x
```

```python
dfOneHot = pd.DataFrame(x, columns = 
                        ["Country_"+str(int(i)) 
                        for i in range(datasets.shape[1]-1)])
df = pd.concat([datasets, dfOneHot], axis=1) #column
#droping the country column
df= df.drop(['Country'], axis=1)
#printing to verify
print(df.head())
```

```python

#4.3 Steps for Handling the missing value

from sklearn.impute import SimpleImputer

datasets=pd.read_csv('/content/drive/MyDrive/SEM-6/Ml/Lab2/Data_for_Missing_Values.csv')
print(datasets)
print(datasets.describe())
```

```python
# Step 3: Seprate Input and Output attributes
# All rows, all columns except last
X = datasets.iloc[:, :-1].values
# Only last column
Y = datasets.iloc[:, -1].values
print("\n\nInput : \n", X)
print("\n\nOutput: \n", Y)
```

```python
# Step 4: Find the missing values and handle it in either way
# 4a. Removing the row with all null values
datasets.dropna(axis=0,how='all',inplace=True)
print(datasets)

#4b. Removing the row with any one null values
#datasets.dropna(axis=0,how='any',inplace=True)
```

```python
updated_df = datasets;
updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df.info()
updated_df['Age'].fillna(updated_df['Age'].mean())
```

```python
imputer = SimpleImputer(missing_values = np.nan,strategy = "mean")
# Fitting the data, function learns the stats
imputer = imputer.fit(X[:, 1:3])
# fit_transform() will execute those stats on the input ie. X[:, 1:3]
X[:, 1:3] = imputer.transform(X[:, 1:3])
# filling the missing value with mean
print("\n\nNew Input with Mean Value for NaN : \n\n", X)
```
